### KIS(Keep It Simple)
- 마찬가지로, 과잉 엔지니어링을 피할 것.
- 최소한의 기능 구현하고 그 이상 복잡하게 하지 말 것.


```python
class ComplicatedNamespace:
    #프로퍼티: 객체의 어떤 속성에 대한 접근 제어하려는 경우 사용
    '''프로퍼티를 가진 객체를 초기화하는 복잡한 예제'''
	ACCEPTED_VALUES = ("id_", "user", "location")

	@classmethod
	def init_with_data(cls, data):
		instance = cls()
		for key, value in data.items():
			if key in cls.ACCEPTED_VALUES:
				setattr(instance, key, value)
		return instance
```

위의 코드를 보면 객체를 초기화 하기위해 추가 클래스 메서드를 만드는 것은 불필요해 보이며, init_with_data라는 일반적이지 않은 메서드의 이름을 알아야 하는 불편함이 있다. 위 상황에서 객체를 초기화 할 때는 init메서드를 사용하는 것이 훨씬 간편할 것이다.

"컴포지션" 활용하여 해결.

컴포지션이란, 상속과 다르게 단순히 사용한다는 개념이다. 즉, 기존의 상속 개념에서의 자식클래스가 부모클래스의 모든 속성을 물려받는게 아니라, 자식클래스가 필요한 속성만 부모클래스로부터 가져와 사용하는 것이다.

일반적으로 상속은 암시적 선언이라고 하며, 컴포지션은 명시적 선언이라고 한다.



출처: https://doorbw.tistory.com/234 [Tigercow.Door]

```python
class TransactionPolicy:
    def __init__(self, policy_data, **extra_data):
        self._data = {**policy_data, **extra_data} #사전을 프라이빗 속성에 저장
    
    def change_in_policy(self, customer_id, **new_policy_data):
        self._data[customer_id].update(**new_policy_data)
        
    def __getitem__(self, customer_id): # getitem으로 slicing을 위한 프록시 생성
        return self._data[customer_id]
    
    def __len__(self):
        return len(self._data)
```

```python
# Composition
# 포함
# 다른 클래스의 일부 기능을 그대로 이용하고 싶으나, 전체 기능 상속은 피하고 싶을 때 사용 Composition 또는 Aggregation 이라고도 한다.
# 상속관계가 복잡할 경우, 코드 이해가 어려운 경우가 많음
# EX) B 에서 A의 객체의 메소드만 가져와서 사용
class Calc:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def add(self):
        return self.x + self.y
 
    def subtract(self):
        return self.x - self.y
 
class Calc2:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def add(self):
        return self.x + self.y
 
    def multiply(self):
        return self.x * self.y

#Calc 클래스에서 Calc2의 multiply() 메서드를 활용하고 싶지만, Calc2 전체를 상속 받고 싶지는 않음
class Calc:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.calc2 = Calc2(x, y) # 해당 클래스의 객체를 명시적으로 가져옴
    
    def add(self):
        return self.x + self.y
 
    def subtract(self):
        return self.x - self.y
    
    def multiply(self):
        return self.calc2.multiply() # 해당 클래스의 객체에 있는 메서드를 명시적으로 활용함
```

메서드 결정 순서(MRO)

- 파이썬에서 다중상속이 발생하면 MRO라는 알고리즘을 사용해 부모 클래스의 메서드 사용 순서를 결정한다.

```python
class A:
    name = "class A" 
    def __init__(self):
        print("class A init")

class B:
    name = "class B"
    def __init__(self):
        print("class B init")

class C(A, B):
    def __init__(self):
        super().__init__()
# B가 가장 상위 클래스이고 A가 그 다음. 메서드 명이 같을 경우 하위 클래스의 것으로 덮어 씌여짐.

C.name 


```

```python
C()
```

위의 코드를 살펴보면 C 클래스는 A, B 두 클래스를 상속받아 두 클래스 모두 갖고 있는 메서드를 실행 했을때 충돌이 발생하지 않고 A 클래스의 메서드가 실행된 것을 확인할 수 있다. MRO 알고리즘을 통해 상속받은 인자의 순서대로 메서드의 우선순위를 갖게 되며, 메서드의 결정 순서는 다음과 같이 확인할 수 있다

```python
[cls.__name__ for cls in C.mro()]
```

믹스인(mixin)
- 코드를 재사용하기 위해 일반적 행동을 캡슐화 해놓은 기본 클래스.
- 그 자체로 유용하지는 않으며 이 클래스만 확장해서는 확실히 동작하지 않음.
- 보통 다른 클래스와 함께 다중 상속하여 믹스인에 있는 메서드 및 속성을 사용함


```python
class BaseTokenizer:
    def __init__(self,str_token):
        self.str_token=str_token
    def __iter__(self):
        yield from self.str_token.split("-")
class UpperIterableMixin:
    def __iter__(self):
        return map(str.upper,super().__iter__()) #부모 클래스가 없는데도 super가 되는구나..?
class Tokenizer(UpperIterableMixin,BaseTokenizer):
    pass

tk=Tokenizer("ABC-EFH")
list(tk)
```

```python
def number_generator():
    x = [1, 2, 3]
    for i in x:
        yield i
 
for i in number_generator():
    print(i)
```

```python
def number_generator():
    x = [1, 2, 3]
    yield from x    # 리스트에 들어있는 요소를 한 개씩 바깥으로 전달
 
for i in number_generator():
    print(i)
```

Mixin 과 Compositon의 차이점
mixin은 IS-A를 의미함.
B IS A (B는 A이다)

Composition은 HAS-A를 의미함.
B HAS A (B는 A를 가지고 있다)

```python
USERS = [(i, f'first_name_{i}', f'last_name_{i}') for i in range(1_000)]

class User:
    def __init__(self, user_id, first_name, last_name):
        self.user_id = user_id
        self.first_name = first_name
        self.last_name = last_name

def users_from_rows(dbrows):
    return [
        User(user_id, first_name, last_name)
        for (user_id, first_name, last_name) in dbrows
    ]

users_from_rows(USERS)
```
